```python
#Problem 1B:
#Sum of n natural numbers
def sumn(inp) :
    ans = 0
    for i in range(1,inp+1) :
        ans = ans + i 
    return ans
#Checking the sum of natural numbers
def sumcheck(inp1 , inp2) :
    if  (inp1 * (inp1 + 1))/2 == inp2:
        return "CORRECT"
```

```python
print(sumn(10))
print(sumcheck(10,sumn(10)))
```

```python
#Sum of squares of n natural numbers
def sumsq(inp) :
    ans = 0
    for i in range(1,inp+1) :
        ans = ans + i * i
    return ans
```

```python
print(sumsq(10))
```

```python
#Problem 2A
def fibo_iteration(inp):
    ar = np.array([1,1])
    for i in range(2,inp+1):
        temp = ar[i-1] + ar[i-2]
        ar = np.append(ar,temp)
    return ar[inp-1]
```

```python
#Problem 2B
def fibo_rec(inp):
    if inp == 1:
        return 1 
    elif inp == 2:
        return 1
    return fibo_rec(inp-1) + fibo_rec(inp-2)
```

```python
def fibo_check(inp1,inp2):
    if inp1==inp2 :
        return "CORRECT: Both the recursive and iterative methods yield the same answer:  " + str(inp1)
```

```python
#Checking both the methods :
print( "Recursion gives  " + str(fibo_rec(10)) )
print( "Iteration gives  " + str(fibo_iteration(10)) )
print ( fibo_check( fibo_rec(10)  ,  fibo_iteration(10)  ) )
```

```python
#Problem 3
#Function for trajectory:
def traj(sx,sy,vx,vy) :
    xval = np.array([])
    yval = np.array([])
    time = 0.001
    ytemp = sy + vy * time - 5 * time * time # Delta Sy = ut + 1/2 a t^2
    while ytemp > 0 :
        yval = np.append(yval,ytemp) 
        xval = np.append(xval,sx + vx * time)  # Delta Sx = ut
        time = time + 0.02
        ytemp = sy + vy * time - 5 * time * time 
    plt.plot(xval,yval,'g.')
    plt.show()
    return 
        
```

```python
#3 A
traj(0,0,2,2)
```

```python
#Problem 4A

# Function for dot product :
def dot(v1,v2):
    return v1[0]*v2[0] + v1[1] * v2[1] + v1[2] * v2[2]

def cycle(si , cv): #cycles index from 0 to 2 based on the argument(Used for cross product and Det Calculation)
    ret = si
    for i in range(cv):
        ret = ret+1
        if(ret==3):
            ret = 0
    #For Debug : print(str(si) + "cycled by" + str(cv) + "is" +str(ret))
    return ret
    
# Function for Cross Product :
def cross(v1,v2):
    ans = np.array([0,0,0]) #Decoy value to specify array size
    for i in range(3):
        ans[i] = v1[cycle(i,1)] * v2[cycle(i,2)] - v1[cycle(i,2)] * v2[cycle(i,1)]
        
    
    return ans
        
    
    
```

```python
v1 = [2,3,4]
v2 = [5,2,1]
print("Dot Product : " + str(dot(v1,v2)))
print("Cross Prodict : " + str(cross(v1,v2)))
```

```python
#Problem 4 B
#MATRIX MULTIPLICATION of 3x3 matrix
mat1 = np.array(np.random.randint(0,10,(3,3)))
mat2 = np.array(np.random.randint(0,10,(3,3)))
print("Matrix 1:")
print(mat1)
print("Matrix 2:")
print(mat2)
```

```python
def matmult(mat1,mat2):
    ans = np.empty((3,3))
    for i in range(3) :
        for j in range(3) :
            temp=0
            for k in range(3):
                temp = temp + mat1[i][k] * mat2[k][j]
            ans[i][j] = temp
    return ans
            
```

```python
print(matmult(mat1,mat2))
```

```python
#Problem 4C
# MATRIX DETERMINANTS
def det(mat1):
    ans = 0
    for i in range(3):
        ans = ans + mat1[0][i]*( mat1[1][cycle(i,1)] * mat1[2][cycle(i,2)] - mat1[1][cycle(i,2)] * mat1[2][cycle(i,1)] )  
    return ans
```

```python
print("The determinant for matrix 1 is :" )
print(det(mat1))
```

```python
#Problem 4D
# Eigen System 
#The Polynomial equation for a matrix:
def polyformer(mat) :
    ans = np.empty(4)
    ans[0] = -1
    ans[1] = mat[0][0] + mat[1][1] + mat[2][2]
    ans[2] = -(mat[0][0]*mat[1][1] + mat[1][1]*mat[2][2] + mat[0][0] * mat[2][2]) + mat[1][2]*mat[2][1] + mat[0][1]*mat[1][0] + mat[0][2]*mat[2][0]
    ans[3] = mat[0][0]*(mat[1][1]*mat[2][2]-mat[1][2]*mat[2][1]) + mat[0][1]*(mat[1][2]*mat[2][0] - mat[1][0]*mat[2][2]) + mat[0][2] * (mat[1][0]*mat[2][1] - mat[1][1]*mat[2][0])
    return ans
```

```python
#The charecteristic equation of matrix 1 : 
#The coefficients of the cubic equation are:
print(polyformer(mat1))
```

```python
#The Eigen Values are:
print("The EigenValues are:")
evalu = np.roots(polyformer(mat1))
print(evalu)
```

```python
def solve(mat,j,k):
    ans =  np.linalg.solve([mat[j][1:],mat[k][1:]],[mat[j][0],mat[k][0]])
    return ans
    
#Function to find Eigen Vector:
def vect(mat,val):
    ans = np.empty(2)
    mat[0][0] = mat[0][0] - val 
    mat[1][1] = mat[1][1] - val 
    mat[2][2] = mat[2][2] - val 
    print(mat)
    #We cant directly solve the linear equation in 3 variables because it will give us the trivial solution of (0,0,0)
    #Therefore we put x=1 in order to find other solution
    #Getting rid of the redundant equation of the three:
    if abs(mat[0][1]/mat[1][1]-mat[0][2]/mat[1][2])<=0.00001:
        ans = solve(mat,1,2)
    elif abs(mat[0][1]/mat[2][1]-mat[0][2]/mat[2][2])<=0.00001:
        ans = solve(mat,1,2)
    elif abs(mat[2][1]/mat[1][1]-mat[2][2]/mat[1][2])<=0.00001:
        ans = solve(mat,3,0)
    else :
        ans = solve(mat,0,1)
        
    return np.append(1,ans)

    
```
