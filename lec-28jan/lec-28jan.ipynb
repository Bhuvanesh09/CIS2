```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
steps=10000
pos=[]
time = np.linspace(0,10,steps)
dt = time[1]-time[0]
cpos = 0
pe = []
ke = []
te = []
m = 1
cp = 10
cv = cp/m
k=10
def force(pos):
    return -k*pos

for i in range(steps) :
    cp = cp + force(cpos) * dt
    cv = cp/m
    cpos = cpos + cv*dt
    pos = np.append(pos,cpos)
    ke = np.append(ke,m*cv*cv/2)
    pe = np.append(pe,k*cpos*cpos/2)
    te = np.append(te,m*cv*cv/2+k*cpos*cpos/2)
    
    
```

```python
plt.plot(time,pos)
```

```python
plt.plot(time,te)
plt.plot(time,pe)
```

```python
plt.plot(time,te)
```
