#### Law of large Numbers:

```python
%matplotlib  notebook
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy
from scipy.special import comb
```

```python
def nCr(n,r):
    return comb(n,r)
```

```python
def generate_array(n,ar):
    output=np.array([]);
    temp = np.random.randint(0,len(ar),size=n)
    for i in temp:
        output = np.append(output,ar[i%len(ar)])
        
    return output;
```

```python
generate_array(5,[-1,1])
```

```python
#Finding average for 100 values:

print(generate_array(100,[-1,1]).sum()/100)
```

```python
def histPlot(n,it,bins):
    avgs = np.array([])
    for i in range(it):
        avgs = np.append(avgs ,generate_array(n,[-1,1]).sum()/n)
    plt.hist(avgs,bins)
    plt.show()
```

```python
histPlot(100,10000,49)
```

```python
def histBinomial(it):
    y = np.array([])    
    for i in range(0,it):
        y = np.append(y , nCr(it,i)/(2**it))
    x = np.linspace(0,1,it)
    #y = y * it
    plt.plot(x,y,label="n = "+str(it))
    plt.legend()
```

```python
histBinomial(100)
histBinomial(500)
histBinomial(1000)
```
