```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[1,2,3],[2,3,3],[3,3,6]])
rand_mat = np.random.randint(0,10,(3,3))
rand_mat
```

```python
val , vect = np.linalg.eig(x)
xaxis=np.array(range(1,4))
val=np.sort(val)
print("Eigenvalues:")
print(val)


print("The Dot Products of the EigenVectors are:")
for i in range(3):
    j = i+1
    if i == 2:
        j=0
    print(int(np.dot(vect[i],vect[j])))
    
```

```python
val=np.sort(val)
col = ['g','r','b']
print(val)
plt.scatter(xaxis,val,marker='o',color=col)
plt.show()
```
